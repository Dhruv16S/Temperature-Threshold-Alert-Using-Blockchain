```python
import pandas as pd
```

```python
data = pd.read_csv("./data.csv")
data
```

```python
print("Dataset information")
data.info()
```

```python
print("Dataset with location as Balta Sound")
filtered_data = data[data["Location"] == "Balta_Sound"]
filtered_data.to_csv("./filtered_data_balta_sound.csv", index = False)
filtered_data
```

```python
print("Data gathered only for the month of June")
filtered_data = filtered_data[filtered_data["Date"] >= "01-06-2023 00:00"]
```

```python
print("Filtered Data Details")
filtered_data.info()
```

```python
filtered_data.to_csv("./filtered_data.csv", index = False)
```

```python
import matplotlib.pyplot as plt
data_points = filtered_data["Temp 2m"]
```

```python
data_points = list(data_points)
plt.figure(figsize = (20, 10))
plt.plot(data_points)
plt.show()
```
